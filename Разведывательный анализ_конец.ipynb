```python
def sud_test(data, column, param):
#Проверяет, есть ли статистичеки значимые различия в переменной param для колонки column.
      
    cols = data.loc[:, column].value_counts().index
    combinations_all = list(combinations(cols, 2))
    for comb in combinations_all:
        if ttest_ind(data.loc[data.loc[:, column] == comb[0], param], 
                     data.loc[data.loc[:, column] == comb[1], param],
                     nan_policy='omit').pvalue \
            <= 0.05/len(combinations_all): # Учли поправку Бонферони
            print('Найдены статистически значимые различия для колонки', column)
            break
            
            
def stat_columns(df_st):
#Функция отображения статистической информации по столбцу дата фрейма.
 
    median = df_st.median()
    mean = df_st.mean()
    std = df_st.std()
    IQR = df_st.quantile(0.75) - df_st.quantile(0.25)
    perc25 = df_st.quantile(0.25)
    perc75 = df_st.quantile(0.75)
    l = perc25 - 1.5*IQR
    r = perc75 + 1.5*IQR
    min_dfst = df_st.min()
    max_dfst = df_st.max()
    # пустые значения
    col_none = pd.isnull(df_st).sum()
    # количество выбросов
    outliers = df_st.loc[(df_st < perc25 - 1.5*IQR) | (df_st> perc75 + 1.5*IQR)]
    # настройки вида диаграммы
    #df_s.hist(label = df_s.name)
    
    range_hist = ((min_dfst - ((max_dfst - min_dfst)/10)), (max_dfst + ((max_dfst - min_dfst)/10)))
    
    outliers.hist(bins = 20, range = range_hist, color = 'red', label = 'выбросы')
    df_st.loc[df_st.between(perc25 - 1.5*IQR, perc75 + 1.5*IQR)].hist(bins = 20,
                                                                    color = 'green',
                                                                    label = df_st.name)
    if min_dfst < l or max_dfst > r:
        return print('Выбросы обнаружены',
                     f'Среднее: {mean}',
                     f'Медиана: {median}',
                     f'Стандартное отклонение: {std}',
                     f'25-й перцентиль: {perc25}',
                     f'75-й перцентиль: {perc75}',
                     f'Интерквартильный интервал(IQR): {IQR}',
                     f'Границы выбросов: {l},{r}',
                     f'Минимальное и максимальное значения: {min_dfst},{max_dfst}',
                     f'Количество выбросов:{len(outliers)}',
                     f'Количество пропусков: {col_none}',
                     {display(outliers)},
                     {plt.legend()}, sep='\n')
    else:
        return print('Выбросы не обнаружены',
                     f'Среднее: {mean}',
                     f'Медиана: {median}',
                     f'Стандартное отклонение: {std}',
                     f'25-й перцентиль: {perc25}',
                     f'75-й перцентиль: {perc75}',
                     f'Интерквартильный интервал(IQR): {IQR}',
                     f'Границы выбросов: {l},{r}',
                     f'Минимальное и максимальное значения: {min_dfst},{max_dfst}',
                     f'Количество пропусков: {col_none}',
                     {plt.legend()}, sep='\n')

    
# Редактирование полей    
def fix_replace(x):
    if pd.isnull(x):
        return x
    if x == 'nan':
        return None
    #x = x.replace(- , '')#studytime_granula,famrel
   #x = x.replace(40, 4) # fedu
    x = x.replace('yes', '1')#schoolsup, famsup,paid,activities,nursery,higher,internet,romantic
    x = x.replace('no', '2')#schoolsup, famsup,paid,activities,nursery,higher,internet,romantic
    x = x.replace('GP', '1')#school
    x = x.replace('MS', '2')#school
    x = x.replace('at_home', '1')#mjob, fjob
    x = x.replace('health', '2')#mjob, fjob
    
    x = x.replace('services', '4')#mjob, fjob
    x = x.replace('teacher', '5')#mjob, fjob
    x = x.replace('course', '1')#reason
    x = x.replace('home', '2')#reason
    x = x.replace('reputation', '4')#reason
    x = x.replace('mother', '1')#studytime_granula,famrel
    x = x.replace('father', '2')#studytime_granula,famrel
    x = x.replace('F', '1')#sex
    x = x.replace('M', '2')#sex
    x = x.replace('U', '1')#address
    x = x.replace('R', '2')#address
    x = x.replace('GT3', '1')#famsize
    x = x.replace('LE3', '2')#famsize
    x = x.replace('A', '1')#pstatus
    x = x.replace('T', '2')#pstatus  
    x = x.replace('other', '3')#mjob, fjob, reason,guardian
    
    return x
```
